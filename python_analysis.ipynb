```python
import pandas as pd

data = pd.read_csv('urb_cpop1_1_Data.csv')
print(pd.__version__) # I ran with 0.23.4
print(len(data))
data.head()
```

```python
# Remove rows that have no value
filtered = data.drop(data[pd.isna(data.Value) | (data.Value == ":")].index)
filtered['key'] = filtered['CITIES'] + ':' + filtered['INDIC_UR']
filtered['Value'] = pd.to_numeric(filtered['Value'])

# pivot(...) will fail because there are some cities (e.g. Barcelona, Bilbao) that have multiple entries
# for the same year
cities = filtered.pivot_table(index='key', columns='TIME', values='Value', aggfunc="mean")
cities.head()
```

```python
# Show biggest cities in 2017

# just 'total' would also give us "Population on the 1st of January, 0-4 years, total" items
key_filter = 'January, total'

cities.filter(like=key_filter,axis=0).sort_values(by=[2017], ascending=False).head(10)
```

```python
# add a growth % 2010 - 2016 column 
cities["growth"] = (cities[2016] / cities[2010] - 1) * 100
cities.filter(like=key_filter,axis=0).sort_values(by=["growth"], ascending=False).head(10)
```
